### Split the dataset into trining and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size =0.8,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

### Import and perform linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

### Predict the chance of admit

```python
y_pred=lr.predict(x_test)
y_pred.round(2)
```

```python
check = pd.DataFrame(x_test,columns=["Weight"])
```

```python
check["Height_Actual"]=y_test
```

```python
check["Height_Predicted"]=y_pred
check
```
